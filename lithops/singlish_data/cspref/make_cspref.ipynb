```python
import pandas as pd
import os
import numpy as np
import json
sample_ratio = 0.3 # ratio to be sampled from all possible pairings
path = 'annotated_cleaned_up/cleaned_up.csv'
np.random.seed(42)
```

```python
df = pd.read_csv(path)
```

```python
# group, pair up and sample
def sample_pairs(group):
    # Get the column names
    cols = group.columns
    
    # Create all pairwise combinations
    pairings = [
        (row1, row2) 
        for idx1, row1 in group.iterrows()
        for idx2, row2 in group.iterrows() 
        if idx1 < idx2
    ]
    
    # Randomly sample of the pairings
    sampled_pairings = np.random.choice(len(pairings), size=int(len(pairings) * sample_ratio), replace=False)
    
    # Create the resulting DataFrame
    result = []
    for idx in sampled_pairings:
        row1, row2 = pairings[idx]
        result.append({
            f'{col}_1': row1[col]
            for col in cols
        } | {
            f'{col}_2': row2[col]
            for col in cols
        })
    
    return pd.DataFrame(result)

# Apply the function to each group
paired_df = df.groupby('src').apply(sample_pairs).reset_index(drop=True)

paired_df
```

```python
def choose_better(metrics):
    score_1 = (
        metrics['human_code-mixedness_1'] * 1 + 
        metrics['human_accuracy_1'] * 1 +
        metrics['human_fluency_1'] * 1
    )
    score_2 = (
        metrics['human_code-mixedness_2'] * 1 + 
        metrics['human_accuracy_2'] * 1 +
        metrics['human_fluency_2'] * 1
    )
    return 'sent_1' if score_1 >= score_2 else 'sent_2'

pref_df = pd.DataFrame({
    'original_l1': paired_df['src_1'],
    'original_l2': paired_df['tgt_1'],
    'sent_1': paired_df['text_1'],
    'sent_2': paired_df['text_2'],
    'chosen': paired_df[['human_code-mixedness_1', 'human_accuracy_1', 'human_fluency_1', 'human_code-mixedness_2', 'human_accuracy_2', 'human_fluency_2']].apply(choose_better, axis=1),
    'lang': 'English-Chinese'
})
pref_df
```

```python
# save
pref_df.to_csv('cspref_zhen.csv', index=False)
```
