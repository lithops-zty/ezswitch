```python
path = 'evaluation_sge-zh.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df_src = df[df.direction=='src']
```

```python
df_tgt = df[df.direction=='tgt']
```

```python
def make_table(df):
    grouped = df.groupby(['method', 'model'])[['GPT_a', 'GPT_f', 'comet', 'bertscore']]
    return grouped.mean()
    
make_table(df_src)
```

```python
make_table(df_tgt)
```
